# ¿En qué consiste el experimento generador del dataset MorphoSem?

El dataset MorphoSem fué generado por un experimento que consiste en el escáner MRI de 40 pacientes siendo 20 neurotípicos y 20 disléxicos. El escáner se realiza durante la realización por parte de los pacientes de una tarea que consiste en leer texto de una pantalla en voz alta.

# Características del dataset



*   20 sujetos neurotípicos
*   20 sujetos disléxicos
*   Tablas descriptoras de las tareas realizadas en las que se reflejan:
  - Estímulos de texto, visuales y auditivos
  - Categoría de la tarea
  - Si requiere respuesta del paciente
  - Duración de la respuesta
  - Marca de tiempo en la que ocurre el evento
  - Marca de ticks (del mri) en la que ocurre el evento
*   Escáneres anatómicos de cada sujeto
*   Escáneres fMRI de cada sujeto durante la realización de las tareas
*   Metadatos de todas las imágenes neurales
*   Tablas descriptoras del desempeño de cada sujeto en sus tareas:
  - Si respondió a la tarea
  - Tiempo de reacción
*   Audios de lectura de cada sujeto
