```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
y, sr = librosa.load('./drive/MyDrive/hola.ogg')
```

```python
librosa.display.waveshow(y, sr=sr)
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
fft = np.fft.fft(y)
magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))
plt.plot(frequency[:len(frequency)//2], magnitude[:len(frequency)//2])
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
```

```python
spec = np.abs(librosa.stft(y, n_fft = 1024, hop_length=256))
spec = librosa.amplitude_to_db(spec, ref=np.max)

librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='log');
plt.colorbar(format='%+2.0f dB');
plt.title('Spectrogram');
```

```python
#MFCC

mfcc= librosa.feature.mfcc(y=y, sr = sr, n_fft=2048, hop_length=512, n_mfcc=30)
librosa.display.specshow(mfcc, sr=sr, x_axis='time', y_axis='log');
plt.colorbar(format='%+2.0f dB');
plt.title('MFCC');
```
